```python
#####################################
# IMPORTING THE LIBRARIES 

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from fbprophet import Prophet
from sklearn.metrics import mean_squared_error, mean_absolute_error
plt.style.use('fivethirtyeight') # For plots
##################################################
```

```python
############################################
# READING THE DATA FROM THE FILES 

test = pd.read_csv('Test_data_forecasting.csv')
train = pd.read_csv('Train_data_forecasting.csv')
validation = pd.read_csv('Validation_test_data.csv')
#############################################
```

```python
#################################################
df = train
# Combining the 2 dataframes 
df = pd.concat([df , validation])
# Creating a column timestap so that data can be fed in fb prophet 
df['timestap'] = df['Date']+" "+df['Time']
# Changing the format to suit FB Prophet 
df['timestap']= pd.to_datetime(df['timestap'] , format='%d/%m/%Y %H.%M.%S')
# Making it as the index 
df =df.set_index('timestap')
##################################################
```

```python
############################################################
#  converting test date and time into timestamp   (everywhere in the code it is misspelled as timestep) 
testf = pd.DataFrame()
testf['timestap'] = test['Date']+" "+test['Time']
testf['timestap']= pd.to_datetime(testf['timestap'] , format='%d/%m/%Y %H.%M.%S')

# as now we have timestamp so dropping date and time column
df.drop(['Date' , 'Time'] , axis=1 , inplace=True)

# Replacing the 0 by O , as it was the only numerical name of the column , rest are all letters 
df = df.rename({'0':'O'}, axis='columns')

# Collecting the list of columns 
clm = list(df.columns)

# Handling the missing values 
for col in clm:
   df[col]= df[col].fillna(method = 'ffill')

#################################################
```

```python
#####################################################
# Importing the required libraries for time series forecasting 
from fbprophet.diagnostics import cross_validation
from fbprophet.diagnostics import performance_metrics

#  mape is storing MAPE for every column and at last i will take average of this
mape=[]

# Iterating throught every column and doing the series forecast 
for col in clm:
    # Making a Dataframe of timestep and individual column and renaming the columns to ds and y as per library requirements 
    subdf = df.reset_index()[['timestap',col]].rename({'timestap':'ds',col:'y'}, axis='columns')
    subdf1 = testf.reset_index()[['timestap']].rename({'timestap':'ds'}, axis='columns')
    # Standardising the data y in the series 
    subdf['y'] = (subdf['y'] - subdf['y'].mean()) / subdf['y'].std()
    # Declaring the model 
    m = Prophet()
    # Training the model 
    m.fit(subdf)
    # Getting the predictions out of the data 
    result = m.predict(subdf1)
    # Just converting 
    s_array = result.yhat.to_numpy()
    # Adding the predictions to a final table 
    testf[col]=s_array
    # Performing cross validation 
    df_cv = cross_validation(m, horizon = '50 hours' , period='20 hours', initial='7900 hours' )
    df_p = performance_metrics(df_cv)
    # Getting the MAPE array 
    mape.append(df_p['mape'].iloc[-1])
    print(df_p.shape)
    print(df_p['mse'].iloc[-1])
    # Plotting it 
    m.plot(result)
#####################################################
```

```python
##################################################################
# Adding the data to the final test dataframe 
for col in clm:
  # Transferring from testf to s_array to test 
  s_array = testf[col].to_numpy()
  test[col]=s_array

# Getting the MAPE Score 
mape = np.array(mape)
MAPE = mape.mean()

# Getting the final predictions 
test.to_csv('ML130_forcasting.csv')
#########################################################
```
